```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
%matplotlib inline
plt.style.use('ggplot')
```

```python
crimes = pd.read_csv('data/crime.csv', encoding = "ISO-8859-1")
```

```python
crimes['OCCURRED_ON_DATE'] = pd.to_datetime(crimes['OCCURRED_ON_DATE'])
```

#### Create subset of noisy party violations

```python
partiescat = ["NOISY PARTY/RADIO-NO ARREST", "NOISY PARTY/RADIO-ARREST"]
```

```python
parties = crimes[crimes.OFFENSE_DESCRIPTION.isin(partiescat)]
```

#### Plot partying in 2016

```python
parties2016 = parties[parties.YEAR == 2016]
```

```python
parties2016 = parties2016.set_index('OCCURRED_ON_DATE')
```

```python
ax = parties2016.resample('W').size().plot(title="Noisy party/radio incidents by week (2016)")
ax.set_xlabel('Week')
```
